```python
import os
DATA_FILE = "bank.txt"

def load_data():
    if not os.path.exists(DATA_FILE):
        with open(DATA_FILE, "w") as file:
            pass
    data = {}
    with open(DATA_FILE, "r") as file:
        for line in file:
            try:
                if line.strip():
                    username, password, bank, balance = line.strip().split(",")
                    data[username] = {
                        "password": password,
                        "bank": bank,
                        "balance": float(balance)
                    }
            except ValueError as e:
                print(f"Error processing line: {line.strip()}. Skipping this line.")
                continue
    return data

def save_data(data):
    with open(DATA_FILE, "w") as file:
        for username, details in data.items():
            line = f"{username},{details['password']},{details['bank']},{details['balance']}\n"
            file.write(line)

def register_user():
    data = load_data()
    username = input("Enter a username: ").strip()
    if username in data:
        print("Username already exists. Try logging in.")
        return
    password = input("Enter a password: ").strip()
    bank = input("Enter bank name (SBI, PNB, Kotak): ").strip().upper()
    data[username] = {
        "password": password,
        "bank": bank,
        "balance": 0.0
    }
    save_data(data)
    print(f"User {username} registered successfully!")

def login_user():
    data = load_data()
    username = input("Enter your username: ").strip()
    if username not in data:
        print("User not found. Please register first.")
        return None
    password = input("Enter your password: ").strip()
    if data[username]["password"] == password:
        print(f"Welcome back, {username}!")
        return username
    else:
        print("Incorrect password.")
        return None
def withdraw(username):
    data = load_data()
    amount = float(input("Enter the amount to withdraw: "))
    if amount > data[username]["balance"]:
        print("Insufficient balance!")
    else:
        data[username]["balance"] -= amount
        save_data(data)
        print(f"Successfully withdrew {amount}. New balance: {data[username]['balance']}")

def deposit(username):
    data = load_data()
    amount = float(input("Enter the amount to deposit: "))
    data[username]["balance"] += amount
    save_data(data)
    print(f"Successfully deposited {amount}. New balance: {data[username]['balance']}")

def check_balance(username):
    data = load_data()
    print(f"Your current balance is: {data[username]['balance']}")

def close_account(username):
    data = load_data()
    confirm = input("Are you sure you want to close your account? (yes/no): ").strip().lower()
    if confirm == "yes":
        del data[username]
        save_data(data)
        print("Your account has been closed.")
        return True
    return False
def main_menu():
    while True:
        print("\n--- Bank Management System ---")
        print("1. Register")
        print("2. Login")
        print("3. Exit")
        choice = input("Choose an option: ").strip()

        if choice == "1":
            register_user()
        elif choice == "2":
            username = login_user()
            if username:
                user_menu(username)
        elif choice == "3":
            print("Exiting the system. Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

def user_menu(username):
    while True:
        print("\n--- User Menu ---")
        print("1. Withdraw Money")
        print("2. Deposit Money")
        print("3. Check Balance")
        print("4. Close Account")
        print("5. Logout")
        choice = input("Choose an option: ").strip()

        if choice == "1":
            withdraw(username)
        elif choice == "2":
            deposit(username)
        elif choice == "3":
            check_balance(username)
        elif choice == "4":
            if close_account(username):
                break
        elif choice == "5":
            print("Logging out...")
            break
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main_menu()
```
